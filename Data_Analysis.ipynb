```python
# Import the libraries you'll need here.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
# Load the csv file into a pandas dataframe.
df = pd.read_csv('data.csv')
sns.set()
plt.figure(figsize=(50,25))
#sdf = df.sort_values(['array_type', 'n'], ascending=True)
c1 = df['array_type'] == "sorted"
sdf = df[c1]
sdf = sdf[['n', 'type', 'execution_time']]
sns_plot = sns.lineplot(x='n', y='execution_time', hue='type', data =sdf)
sns_plot.figure.savefig("sorted_execution_time.png")
```
